## Matplotlib

-Used for 2D & 3D visualization

```python
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
import numpy as np
x = np.linspace(10,100 , 50)
y = np.log(x)
```

```python
print(x)
print(y)
```

### Functional method vs Object oriented method

#### FUNCTIONAL

```python
plt.plot(x, y, 'violet') 
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('X-Y PLANE')
plt.show()
```

**to create subplots**

```python
plt.subplot(2,3,1) # SYNTAX: (no. of rows, no. of columns, plot no.) # plot1
plt.plot(x, y, 'green') # More on color options later

plt.subplot(2,3,2) # plot2
plt.plot(y, x, 'blue')

plt.subplot(2,3,3) # plot3
plt.plot(x+y, y, 'red')

plt.subplot(2,3,4) # plot4
plt.plot(x*y, y, 'yellow')

plt.subplot(2,3,5) # plot5
plt.plot(y, x*y, 'black')

plt.subplot(2,3,6)# plot6
plt.plot(x/y, x+y, 'grey')
```

#### OBJECT ORIENTED 

```python
'''
Just to compare the functional method & object oriented method

plt.plot(x, y, 'violet') 
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('X-Y PLANE')
plt.show()

'''

# empty canvas 
fig = plt.figure()

# adding set of axes - MUST ADD AXES... 
axes = fig.add_axes([1, 1, 1, 1]) # left, bottom, width, height (range 0 to 1)

# plotting 
axes.plot(x, y, 'y') # means yellow 
# ~ plt.figure().add_axes([1,1,1,1]).plot(x,y,'r')
axes.set_xlabel('X AXIS')
axes.set_ylabel('Y AXIS')
axes.set_title('X-Y PLANE')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.2, 0.15, 1, 1]) # main axes
ax2 = fig.add_axes([0.3, 0.8, 0.3, 0.3]) # inset axes

# fig 1
ax1.plot(x, y, 'y')
ax1.set_title('Axes 1')

# fig 2
ax2.plot(y, x, 'g')
ax2.set_title('Axes 2');
```

```python
'''

# subplot 

plt.subplot(2,3,1)
plt.plot(x, y, 'green')
plt.subplot(2,3,2) # plot2
plt.plot(y, x, 'blue')
....... so on 

'''

x = np.linspace(10,100 , 50)
y = np.log(x)
z = np.exp(x)
# for subplots 

fig, axes = plt.subplots(nrows=1, ncols=2)

#axes, since 2 plots... 
axes[0].plot(x, y, 'g', lw=3, ls='--') # line width & line style
axes[1].plot(x,z,'purple', lw=1, ls='solid')

fig
plt.tight_layout() # to adjust overlapping subplots
```

```python
fig = plt.figure(figsize=(5,4), dpi=120) # dpi is dots/pixels per inch 
# 5*4 is the size...
axes = fig.add_axes([1,1,1,1])
axes.plot(x,y,'y')

# we can set the xlabel, ylabel, set_title
```

```python
# to save the file: fig.savefig("graph.png", dpi=200) # you can specify the dpi else ignore it
```

```python
# to add legends
x = np.linspace(10,100,50)
y = x/2
z = x/3

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y,'red', label="y: x/2")
ax.plot(x, z,'blue', label="z: x/3")
ax.legend(loc=2) # we can have different loc values; 1,2,3,4.
```

```python
# different plots... 
fig, ax = plt.subplots(figsize=(10,5))

'''
alpha - sets the opacity of line
linewidth or lw - width of line
'''

ax.plot(x, x+1, color="blue", linewidth=0.5)

ax.plot(x, x+5, color="green", lw=1, linestyle='-')

ax.plot(x, x+10, color="black", lw=3, ls='-', marker='+')

ax.plot(x, x+15, color="yellow", lw=1, ls='-', marker='o', markersize=2)

ax.plot(x, x+20, color="pink", lw=1, ls='-', marker='s', markersize=8,
markerfacecolor="orange", markeredgewidth=3, markeredgecolor="violet")
```
