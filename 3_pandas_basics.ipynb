```python
# similarly to drop rows: 
df.drop('A', axis=0, inplace=True) # to actually drop the row...
print(df)
```

### dealing with missing data with pandas

```python
df = pd.DataFrame({
     'A':[10,20,np.nan,43,61,np.nan,92,np.nan],
     'B':[53,23,np.nan,51,77,62,np.nan,85],
     'C':[9,8,7,6,5,4,3,2]
})
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)   # to drop columns with NA values 
```

```python
df.dropna(thresh=2) #minimum amount of NA values to drop
```

```python
df.fillna(value=0) # filling with 0
```

```python
df['C'].fillna(value=df['C'].var()) # filling with var value...
```

```python
# groupby used to group rows together and perform aggregate functions
x = df.groupby('C')
x  # object saved
```

```python
x.mean()
# similarly std, var, min, max, count, transpose()...etc.
```

```python
x.describe()
```

#### joining & concatention...

```python


t1 = pd.DataFrame({'A': ['earth', 'venus', 'mars', 'pluto'],
                     'B': ['green', 'red', 'blue', 'violet'],
                     'C': ['pen', 'pencil', 'eraser', 'sharpner']})
   
t2 = pd.DataFrame({'Q': ['100', '200', '300', '400'],
                          'W': ['1', '2', '3', '4'],
                          'E': ['5', '6', '7', '8']})    

print(t1)
print(t2)
```

```python
pd.concat([t1,t2])
```

```python
pd.concat([t1,t2], axis=1)
```

```python
#JUST LIKE SQL JOIN
#pd.merge()

print(t1.join(t2))
print('')
print(t1.join(t2, how = 'inner'))
```
