## plotly & cufflinks
- It's a graphic & analytics tool.
- We can create these plots: scatter, bar, box, spread, ratio, heatmap, surface, histogram, bubble..... 

```python
import pandas as pd
import numpy as np
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
```

```python
init_notebook_mode(connected=True) # this would help plotly connect python to an interactive js library....
cf.go_offline() # offline use...
```

```python
data = pd.DataFrame(np.random.randn(1000,6),columns='A B C D E F'.split())
data.head(3)
```

```python
data.iplot(kind='scatter',x='A',y='B',mode='markers',size=5)
# we can specify other kinds like box/heatmap... etc..., & specify the X & Y if necessary.

# other example: histogram... 
data['F'].iplot(kind='hist',bins=25)
```
